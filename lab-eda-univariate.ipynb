```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")  
```

```python
df.head()
```

```python
# PART 1: UNDERSTANDING PRODUCT CATEGORIES

# 1.1 Frequency Tables
category_counts = df['category'].value_counts()
top_5_categories = category_counts.head(5)
top_5_categories
```

```python
# 1.2 Visualizations
category_counts.plot(kind='bar')
plt.title('Distribution of Products Across Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

```python
# 1.2 Visualizations
top_5_categories.plot(kind='bar')
plt.title('Distribution of Products Across Top 5 Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

```python
# 1.2 Visualizations
top_5_categories.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 Product Categories by Proportion')
plt.ylabel('')
plt.show()
```

```python
# PART 2: DELVING INTO PRODUCT PRICING

#2.1 Measures of Centrality
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print("\nPrice Centrality Measures:")
print(f"Mean Price: {mean_price:.2f}")
print(f"Median Price: {median_price:.2f}")
print(f"Mode Price: {mode_price:.2f}")
```

```python
# 2.2 Measures of Dispersion
variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print("\nPrice Dispersion Measures:")
print(f"Variance: {variance_price:.2f}")
print(f"Standard Deviation: {std_dev_price:.2f}")
print(f"Range: {range_price:.2f}")
print(f"Interquartile Range: {iqr_price:.2f}")
```

```python
# 2.3 Visualizations
df['price'].plot(kind='hist', bins=50)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# 2.3 Visualizations
df.boxplot(column='price')
plt.title('Box Plot of Product Prices')
plt.ylabel('Price')
plt.show()
```

```python
# PART 3: UNPACKING PRODUCT RATINGS

#3.1 Measures of Centrality
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print("\nRating Centrality Measures:")
print(f"Mean Rating: {mean_rating:.2f}")
print(f"Median Rating: {median_rating:.2f}")
print(f"Mode Rating: {mode_rating:.2f}")
```

```python
# 3.2 Measures of Dispersion
variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()
iqr_rating = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print("\nRating Dispersion Measures:")
print(f"Variance: {variance_rating:.2f}")
print(f"Standard Deviation: {std_dev_rating:.2f}")
print(f"Interquartile Range: {iqr_rating:.2f}")
```

```python
# 3.3 Shape of the Distribution
skewness = df['stars'].skew()
kurtosis = df['stars'].kurt()

print("\nShape of the Rating Distribution:")
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

```python
# 3.4 Visualizations
df['stars'].plot(kind='hist', bins=10)
plt.title('Distribution of Product Ratings')
plt.xlabel('Stars')
plt.show()
```
